### Cleaning Data

Mengubah format kolom "dteday" pada dataframe day_df dan hour_df menjadi format datetime

```python
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
```

Menyamakan data "weathersit" pada kedua dataframe

```python
#Membuat dataframe yang hanya tanggal dan informasi cuaca
date_weather_hour_df = hour_df.groupby(by="dteday").agg({"weathersit": "max"})
date_weather_hour_df.sort_values(by="weathersit", ascending=False).head(12)

#Membuat dataframe dari day_df tanpa kolom weathersit
weatherless_day_df = day_df.drop("weathersit", axis=1)

#Menggabungkan kedua dataframe baru
new_day_df = pd.merge(
    left=weatherless_day_df,
    right=date_weather_hour_df,
    how="outer",
    left_on="dteday",
    right_on="dteday"
)
new_day_df.head()
```

Drop beberap kolom yang dianggap tak perlu

```python
new_day_df.drop(columns=['holiday', 'workingday', 'temp', 'atemp', 'hum', 'windspeed'])
```

Menjadikan dataframe "new_day_df" terpisah sesuai tahunnya

```python
eleven_day_df = new_day_df.drop(new_day_df[new_day_df["yr"]==1].index)
twelve_day_df = new_day_df.drop(new_day_df[new_day_df["yr"]==0].index)
```

## Exploratory Data Analysis (EDA)

### Explore eleven_day_df

Membuat dataframe baru jumlah user rental bike tiap bulannya

```python
monthly_users_eleven_day_df = eleven_day_df.resample(rule="M", on="dteday").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
})
monthly_users_eleven_day_df.index = monthly_users_eleven_day_df.index.strftime('%B')
monthly_users_eleven_day_df.reset_index()
```

Membuat dataframe baru tentang jumlah user tiap season

```python
season_users_eleven_day_df = eleven_day_df.groupby(by="season").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
}).reset_index()
season_users_eleven_day_df.season.replace([1, 2, 3, 4], ["spring", "summer", "autumn", "winter"], inplace=True)
season_users_eleven_day_df.head()
```

Membuat dataframe baru tentang jumlah user tiap weather

```python
weather_users_eleven_day_df = eleven_day_df.groupby(by="weathersit").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
}).reset_index()
weather_users_eleven_day_df.replace([1, 2, 3, 4], ["clear", "mist", "slightly bad", "very bad"], inplace=True)
weather_users_eleven_day_df.head(4)
```

### Explore twelve_day_df

Membuat dataframe baru jumlah user rental bike tiap bulannya

```python
monthly_users_twelve_day_df = twelve_day_df.resample(rule="M", on="dteday").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
})
monthly_users_twelve_day_df.index = monthly_users_twelve_day_df.index.strftime('%B')
monthly_users_twelve_day_df.reset_index()
```

Membuat dataframe baru tentang jumlah user tiap season

```python
season_users_twelve_day_df = twelve_day_df.groupby(by="season").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
}).reset_index()
season_users_twelve_day_df.season.replace([1, 2, 3, 4], ["spring", "summer", "autumn", "winter"], inplace=True)
season_users_twelve_day_df.head()
```

Membuat dataframe baru tentang jumlah user tiap weather

```python
weather_users_twelve_day_df = twelve_day_df.groupby(by="weathersit").agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
}).reset_index()
weather_users_twelve_day_df.weathersit.replace([1, 2, 3, 4], ['clear', 'mist', 'slightly bad', 'very bad'], inplace=True)
weather_users_twelve_day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perkembangan rental bike dari data dua tahun tersebut?

Membuat line chart untuk penggunaan rental bike pada tahun 2011 dan 2012 pada tiap bulannya

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

sns.lineplot(x="dteday", y="cnt", data=monthly_users_eleven_day_df, orient='x', ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Jumlah User per Bulan Tahun 2011", loc="center", fontsize=17)
ax[0].tick_params(axis ='y', labelsize=12)

sns.lineplot(x="dteday", y="cnt", data=monthly_users_twelve_day_df, orient='x', ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Jumlah User per Bulan Tahun 2012", loc="center", fontsize=17)
ax[1].tick_params(axis ='y', labelsize=12)
```

### Pertanyaan 2: Bagaimana pengaruh cuaca dan musim terhadap bisnis bike rental?

Membuat grafik user pada saat  musim tertentu di tahun 2011

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
season_colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
weather_colors = ["#D3D3D3", "#72BCD4", "#D3D3D3", "#D3D3D3"]

sns.lineplot(x="dteday", y="cnt", data=monthly_users_eleven_day_df, orient='x', ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)
ax[0].set_title("Jumlah User per Bulan Tahun 2011", loc="center", fontsize=17)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="season", y="cnt", data=season_users_eleven_day_df,palette=season_colors, hue='season', ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("Jumlah User per Musim Tahun 2011", loc="center", fontsize=17)
ax[1].tick_params(axis ='y', labelsize=12)

sns.barplot(x="weathersit", y="cnt", data=weather_users_eleven_day_df, palette=weather_colors,hue='weathersit', ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("Jumlah User Pada Tiap Cuaca di Tahun 2011", loc="center", fontsize=17)
ax[2].tick_params(axis ='y', labelsize=12)
```

Membuat grafik user pada saat  musim tertentu di tahun 2012

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
season_colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
weather_colors = ["#D3D3D3", "#72BCD4", "#D3D3D3", "#D3D3D3"]

sns.lineplot(x="dteday", y="cnt", data=monthly_users_twelve_day_df, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=45)
ax[0].set_title("Jumlah User per Month di 2012")
ax[0].tick_params(axis='y', labelsize=12)

sns.barplot(x="season", y="cnt", data=season_users_twelve_day_df, palette=season_colors, hue='season', ax=ax[1])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Jumlah User per Musim di 2012")
ax[0].tick_params(axis='y', labelsize=12)

sns.barplot(x="weathersit", y="cnt", data=weather_users_twelve_day_df, palette=weather_colors, hue='weathersit', ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("Jumlah User Pada Tiap Cuaca di Tahun 2012", loc="center", fontsize=17)
ax[2].tick_params(axis ='y', labelsize=12)
```

## Conclusion

- Conclusion :
    Pada tahun 2011, Perkenbangan user rental bike mengalami kenaikan pada bulan Maret dan April yang mana user rental bike terjadi pada bulan Juni. Setelah itu, user rental bike di bulan - bulan kedepannya kurang lebih sama hingga mengalami penurunan pada bulan Oktober hingga Desember

##### Pertanyaan 2: Bagaimana pengaruh cuaca dan musim terhadap bisnis bike rental?
